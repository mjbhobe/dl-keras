## About the Fashion-MNIST dataset. 
Fashion-MNIST can be used as drop-in replacement for the original MNIST dataset (of 10 categories of handwritten digits). It shares the same image size (28x28 graysacle) and has 60,000 training and 10,000 testing splits. However, unlike the 'simple' number images, the images from this dataset are a bit involved. So, our CNN network architecture will be different.

Per [Zolando](https://github.com/zalandoresearch/fashion-mnist), the Fashion-MNIST dataset was created as a replacement for the MNIST dataset because:
* **MNIST is too easy**. Convolutional nets can achieve 99.7% on MNIST. Classic machine learning algorithms can also achieve 97% easily. 
* **MNIST is overused**. 
* **MNIST cannot represent modern CV tasks**
Take a look at the link above and read the various articles it references.

We have 10 categories of output labels - unlike the MNIST dataset, where they were values for numbers, we have distinct types here.

Label Description 
* 0 - T-shirt/top 
* 1 - Trouser 
* 2 - Pullover 
* 3 - Dress 
* 4 - Coat 
* 5 - Sandal 
* 6 - Shirt 
* 7 - Sneaker 
* 8 - Bag 
* 9 - Ankle boot

In this notebook, we will develop a Convolutional Neural Network (CNN) using Tensorflow 2.0 with a Keras front-end. We'll use Tensorflow's implementation of Keras (viz. `tf.keras`). We'll progressively _improve_ the CNN using regularization and Image Augmentation.

## Loading Data
* We'll load the data from `tf.keras.datasets.fashion_mnist` package. 
* Next, we'll split the _test_ dataset into _cross-validation_ and _test_ sets in `80:20` ratio.
* We'll normalize the images, so that all values are between `0.0` and `1.0`

Why split the `test` dataset and not the `train` dataset?
* Prof. Andrew Ng advises, in his `Structured ML` course on Coursera, that the cross-val and test datasets should be drawn from the same population! Hence we split the `test` dataset.

```python
def load_data(debug=False):
    from tensorflow.keras.datasets.fashion_mnist import load_data
    
    (X_train, y_train), (X_test, y_test) = load_data()
    if debug:
        print(f"After loading...")
        print(f"   - X_train.shape = {X_train.shape}, y_train.shape = {y_train.shape}, " +
              f"X_test.shape = {X_test.shape}, y_train.shape = {y_train.shape}")
        
    # shuffle the datasets, to get rid of any implicit sorting
    indexes = np.random.permutation(np.arange(X_train.shape[0]))
    X_train, y_train = X_train[indexes], y_train[indexes]
    indexes = np.random.permutation(np.arange(X_test.shape[0]))
    X_test, y_test = X_test[indexes], y_test[indexes]
    
    # split X_test into cross-val & test sets in 80:20 
    num_recs = int(0.80 * X_test.shape[0])
    X_val, y_val = X_test[:num_recs], y_test[:num_recs]
    X_test, y_test = X_test[num_recs:], y_test[num_recs:]
    if debug:
        print(f"After splitting...")
        print(f"   - X_train.shape = {X_train.shape}, y_train.shape = {y_train.shape}, " +
              f"X_val.shape = {X_val.shape}, y_val.shape = {y_val.shape}, " +
              f"X_test.shape = {X_test.shape}, y_test.shape = {y_test.shape}")
    
    # normalize the images
    X_train = X_train.astype('float32') / 255.0
    X_val = X_val.astype('float32') / 255.0
    X_test = X_test.astype('float32') / 255.0
    
    # resize images
    X_train = np.expand_dims(X_train, axis=3)
    X_val = np.expand_dims(X_val, axis=3)
    X_test = np.expand_dims(X_test, axis=3)
    
    if debug:
        print(f"Before returning...")
        print(f"   - X_train.shape = {X_train.shape}, y_train.shape = {y_train.shape}, " +
              f"X_val.shape = {X_val.shape}, y_val.shape = {y_val.shape}, " +
              f"X_test.shape = {X_test.shape}, y_test.shape = {y_test.shape}")
    
    return (X_train, y_train), (X_val, y_val), (X_test, y_test)
```

```python
def display_sample(sample_images, sample_labels, sample_predictions=None, num_rows=5, num_cols=10,
                   plot_title=None, fig_size=None):
    
    import seaborn as sns
  
    """ display a random selection of images & corresponding labels, optionally with predictions
        The display is laid out in a grid of num_rows x num_col cells
        If sample_predictions are provided, then each cell's title displays the prediction (if it matches
        actual)
        or actual/prediction if there is a mismatch
    """
    assert sample_images.shape[0] == num_rows * num_cols
    LABELS = {
        0: 'T-shirt/Top',
        1: 'Trouser',
        2: 'Pullover',
        3: 'Dress',
        4: 'Coat',
        5: 'Sandal',
        6: 'Shirt',
        7: 'Sneaker',
        8: 'Bag',
        9: 'Ankle Boot'
    }

    with sns.axes_style("whitegrid"):
        sns.set_context("notebook", font_scale=1.1)
        sns.set_style({"font.sans-serif": ["Verdana", "Arial", "Calibri", "DejaVu Sans"]})

        f, ax = plt.subplots(num_rows, num_cols, figsize=((14, 10) if fig_size is None else fig_size),
            gridspec_kw={"wspace": 0.05, "hspace": 0.35}, squeeze=True)

        for r in range(num_rows):
            for c in range(num_cols):
                image_index = r * num_cols + c
                ax[r, c].axis("off")
                # show selected image
                sample_image = np.squeeze(sample_images[image_index], axis=2)
                ax[r, c].imshow(sample_image, cmap="Greys", interpolation='nearest')

                if sample_predictions is None:
                    # show the actual labels in the cell title
                    title = ax[r, c].set_title(f"{LABELS[sample_labels[image_index]]}")
                else:
                    # else check if prediction matches actual value
                    true_label = LABELS[sample_labels[image_index]]
                    pred_label = LABELS[sample_predictions[image_index]]
                    prediction_matches_true = (true_label == pred_label)
                    if prediction_matches_true:
                        # if actual == prediction, cell title is prediction shown in green font
                        title = f"{true_label}"
                        title_color = 'g'
                    else:
                        # if actual != prediction, cell title is actua/prediction in red font
                        title = f"{true_label}/{pred_label}"
                        title_color = 'r'
                    # display cell title
                    title = ax[r, c].set_title(title)
                    plt.setp(title, color=title_color)
        # set plot title, if one specified
        if plot_title is not None:
            f.suptitle(plot_title)

        plt.show()
        plt.close()
```

```python
(X_train, y_train), (X_val, y_val), (X_test, y_test) = \
    load_data(debug=True)
```

```python
# display a sample of the loaded dataset
SAMPLE_SIZE = 50
rand_indexes = np.random.randint(0, X_test.shape[0], SAMPLE_SIZE)
sample_images = X_test[rand_indexes]
sample_labels = y_test[rand_indexes]
display_sample(sample_images, sample_labels, plot_title=f'Sample of {SAMPLE_SIZE} images')
```

## Building The Model
In this section, we'll develop a base model, which we will progressively _improve_ using regularization and Image Augmentation.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
import kr_helper_funcs as kru
```

```python
# global hyper parameters
IMAGE_HEIGHT, IMAGE_WIDTH, NUM_CHANNELS, NUM_CLASSES = 28, 28, 1, 10
NUM_EPOCHS, BATCH_SIZE, LEARNING_RATE, L2_REG = 25, 32, 0.001, 0.0005
MODEL_SAVE_DIR = os.path.join('.', 'model_states')
```

### Base Model (no regularization)

We'll build the model using the Keras `Sequential` API, with 3 `Dense` layers. We'll use the `Adam` optimizer to optimize a `sparse_categorical_crossentropy` losss. We will also track `sparse_categorical_accuracy` metric apart from the `loss` metric.

**NOTE**: 
* We **did not** `one-hot-encode` the class labels. Hence we use the `sparse_categorical_crossentropy` loss metric.
* Since our images are all 3D shaped `(28, 28, 1)`, hence we have flattened the images using the `Flatten` layer before feeding to the `Dense` layers.

```python
# build the base model - no regularization
def build_model():
    model = Sequential([
        Conv2D(256, kernel_size=(5,5), padding='SAME', activation='relu',
               input_shape=(IMAGE_HEIGHT, IMAGE_WIDTH, NUM_CHANNELS)),
        MaxPooling2D((2, 2)),
        Conv2D(128, kernel_size=(5,5), padding='SAME', activation='relu'),
        MaxPooling2D((2, 2)),
        Conv2D(64, kernel_size=(3,3), padding='SAME', activation='relu'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(512, activation='relu'),
        Dense(NUM_CLASSES, activation='softmax')
    ])
    opt = Adam(learning_rate=LEARNING_RATE, decay=0.005)
    model.compile(optimizer=opt, loss='sparse_categorical_crossentropy',
                  metrics=['sparse_categorical_accuracy'])
    return model
```

```python
model = build_model()
print(model.summary())
```

```python
hist = model.fit(X_train, y_train, epochs=NUM_EPOCHS, batch_size=BATCH_SIZE,
                 validation_data=(X_val, y_val))
```

```python
kru.show_plots(hist.history, metric='sparse_categorical_accuracy')
```

```python
loss, acc = model.evaluate(X_train, y_train, verbose=0)
print(f"Training data -> loss: {loss:.4f}, acc: {acc:.4f}")
loss, acc = model.evaluate(X_val, y_val, verbose=0)
print(f"Cross-val data -> loss: {loss:.4f}, acc: {acc:.4f}")
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print(f"Test data -> loss: {loss:.4f}, acc: {acc:.4f}")
```

**Observations:**

>Model | Training Acc | Cross-Val Acc | Testing Acc 
>:---|:---:|:---:|:---:|
>**Keras Base Model**|99.06%|97.94%|98.20%|

* The model gives us a test accuracy of `~98%`, which is pretty good!
* There is `~2%` difference between `training accuracy` (`99.06%`) and `Cross-val accuracy` (`97.94%`), which implies that there is _some_ overfitting of the model to training data. 
    * Perhaps a greater value for `L2_REG` or `Dropout` percentage should improve this (i.e. reduce the difference)
* There is not much difference between `Cross-val accuracy` (`97.94%`) and `Test accuracy` (`98.20%`) - this should be the case as both cross-val and test datasets came from the same *population*
